```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.read_csv("D:/Datasets/Mobilephone_data_final - ndtv_data_final.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

### Scatter plot - Resolution x Vs Resolution y 

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Resolution x', y='Resolution y', data=df)
plt.title('Resolution x Vs Resolution y', fontweight='bold')
```

### Number of Mobiles by Brand 

```python
x=df.groupby('Brand').count().reset_index()['Name']
y=df.groupby('Brand').count().reset_index()['Brand']
plt.figure(figsize=(20,5))
sns.barplot(x=y, y=x)
plt.xticks(rotation=90)
plt.title('Number of Mobiles by Brand', fontweight='bold')
```

### Price by Brand

```python
x=df.groupby('Brand').Price.median().reset_index()['Price']
y=df.groupby('Brand').Price.median().reset_index()['Brand']
plt.figure(figsize=(20,5))
sns.barplot(x=y, y=x)
plt.xticks(rotation=90)
plt.title('Price by Brand', fontweight='bold')
```

### Box plot chart of price

```python
sns.boxplot(x=df['Price'])
```

### Scatter chart between 'Battery capacity (mAh)' and 'Price'. 

```python
plt.figure(figsize=(8,6))
plt.scatter(x=df['Battery capacity (mAh)'],y=df['Price'], c=df['Number of SIMs'])
```

### Correlation

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True)
```

### Distribution chart for price

```python
plt.figure(figsize=(8,6))
sns.displot(x=df['Price'])
```
